### Draw a plot representing costing increase

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datasets_count = 1
```

```python
plt.figure(figsize=(32, 18))
for i in range(datasets_count):
    df = pd.read_csv("dataset_full" + str(i) + ".csv", header=None)
    df.rename(columns = {0: "funders", 1: "cost", 2: "proof"}, inplace=True)
    plt.plot(df["funders"], df["cost"], label="Cost", color="green")
    plt.scatter(df["funders"], df["cost"], color="blue", alpha="0.15")
    
plt.title("Cost function by funders count")
plt.xlim((0, 200))
plt.ylim((0, 10000))
# plt.legend()
plt.grid(ls=":")
plt.savefig("outp-full.png")
plt.show();
```
